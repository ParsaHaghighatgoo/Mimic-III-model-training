```python
import pandas as pd
from datetime import datetime

# Load the CSV files
path = "J:\\mimic-iii-clinical-database-1.4\\mimic-iii-clinical-database-1.4"
# load Patients and Diagnoses_icd
patients = pd.read_csv(path + '\\patients.csv')
admissions = pd.read_csv(path + '\\admissions.csv')

# Read the CSV files into dataframes
df1 = pd.read_csv(patients)
df2 = pd.read_csv(admissions)

# Ensure both files have DOB and ADMITTIME columns
if 'DOB' in df1.columns and 'ADMITTIME' in df2.columns:
    # Combine the two dataframes
    combined_df = pd.merge(df1, df2, on='common_column')  # Replace with actual common column if needed

    # Define a function to extract the year using your code logic
    def extract_year(date_str):
        try:
            # Parsing the date string
            date_object = datetime.strptime(date_str, "%m/%d/%Y %H:%M")
            # Extracting the year
            return date_object.year
        except ValueError:
            return None  # Handle invalid date formats

    # Apply the function to extract years from DOB and ADMITTIME
    combined_df['DOB_Year'] = combined_df['DOB'].apply(extract_year)
    combined_df['ADMIT_Year'] = combined_df['ADMITTIME'].apply(extract_year)

    # Calculate the difference between ADMIT_Year and DOB_Year
    combined_df['Year_Difference'] = combined_df['ADMIT_Year'] - combined_df['DOB_Year']

    # Display the resulting dataframe
    print(combined_df[['DOB', 'ADMITTIME', 'Year_Difference']])
    
    # Optionally save to a new CSV
    combined_df.to_csv('output.csv', index=False)
else:
    print("Columns 'DOB' and 'ADMITTIME' are not present in both files.")
```
